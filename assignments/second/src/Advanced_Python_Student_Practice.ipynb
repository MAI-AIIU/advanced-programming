## Activity 1 — Use the iteration protocol

Complete the function using `iter()`, `next()`, and `StopIteration`.

> Work independently first, then compare approaches with a partner.

```python
def consume(iterable):
    collected = []
    iter_ = iter(iterable)

    while True:
        try:
            collected.append(next(iter_))
        except StopIteration:
            break
    return collected

print(consume([4, 8, 12]))
# Expected: [4, 8, 12]
```

## Activity 2 — Create a custom iterator

Complete `BatchNumbers` so it produces batch numbers from 1 through `maximum`.

> Work independently first, then compare approaches with a partner.

```python
class BatchNumbers:
    def __init__(self, max):
        self.current = 1
        self.max = max

    def __iter__(self):
        return self

    def __next__(self):
        if self.current > self.max:
            raise StopIteration

        value = self.current
        self.current += 1
        return value

print(list(BatchNumbers(5)))
```

## Activity 6 — Build a sensor pipeline

Keep numeric heart rates from 20 to 250, then normalize them by dividing by 250.

> Work independently first, then compare approaches with a partner.

```python
raw_heart_rates = [72, None, 95, -5, 280, 80, "error", 110]

def keep_valid(values):
    for value in values:
        if isinstance(value, (int, float)) and 0 <= value <= 220:
            yield value

def normalize(values): # the best approach to normalize is (number - min / max - min)
    for value in values:
        yield value / 100

pipeline = normalize(keep_valid(raw_heart_rates))
print(list(pipeline))
```

## Activity 7 — Create mini-batches

Yield full batches and optionally keep the incomplete final batch.

> Work independently first, then compare approaches with a partner.

```python
def create_batches(iter_, batch_size, drop_last=False):

    batch = []
    for item in iter_:
        batch.append(item)
        if len(batch) == batch_size:
            yield batch
            batch = []

    if batch and not drop_last:
        yield batch


print(list(create_batches(range(10), 3)))
print(list(create_batches(range(10), 3, True)))
```

# Part D — Final AI Data-Loader Challenge

An elderly-care dataset contains age, daily steps, heart rate, and a risk label. Build a lazy loader that parses, validates, transforms, and batches the data. It must create a fresh pipeline for every epoch.

```python
from pathlib import Path
import csv

dataset_path = Path("ai_practice.csv")

rows = [
    ["age", "steps", "heart_rate", "label"],
    [65, 3200, 78, "normal"],
    [72, 1400, 91, "risk"],
    [68, 5100, 74, "normal"],
    [75, 800, 105, "risk"],
    [70, 4300, 80, "normal"],
    [200, 1000, 70, "normal"],
    [69, "unknown", 76, "normal"],
    [80, 600, 110, "risk"],
]

with dataset_path.open("w", newline="", encoding="utf-8") as file:
    csv.writer(file).writerows(rows)

print(dataset_path)
```

## Complete the loader

Implement all TODO sections. Invalid rows should be reported rather than silently discarded.

```python
from pathlib import Path
import csv

def read_csv_records(path):
    with Path(path).open("r", newline="", encoding="utf-8") as file:
        yield from csv.DictReader(file)


def parse_records(records):
    for row in records:
        try:
            row["age"] = int(row["age"])
            row["steps"] = int(row["steps"])
            row["heart_rate"] = int(row["heart_rate"])
            yield row
        except (ValueError, TypeError):
            print("Skipping invalid row:", row)


def validate_records(records):
    for row in records:
        if not (60 <= row["age"] <= 120):
            print("Invalid age:", row)
            continue
        if row["steps"] < 0:
            print("Invalid steps:", row)
            continue
        if not (30 <= row["heart_rate"] <= 220):
            print("Invalid heart rate:", row)
            continue
        if row["label"] not in ("normal", "risk"):
            print("Invalid label:", row)
            continue

        yield row


def transform_records(records):
    for row in records:
        yield {
            "age": row["age"] / 100,
            "steps": row["steps"] / 10000,
            "heart_rate": row["heart_rate"] / 220,
            "label": 0 if row["label"] == "normal" else 1,
        }


def create_batches(iterable, batch_size):
    batch = []
    for item in iterable:
        batch.append(item)
        if len(batch) == batch_size:
            yield batch
            batch = []
    if batch:
        yield batch


class PracticeDataLoader:
    def __init__(self, path, batch_size):
        self.path = path
        self.batch_size = batch_size

    def __iter__(self):
        records = read_csv_records(self.path)
        records = parse_records(records)
        records = validate_records(records)
        records = transform_records(records)
        return create_batches(records, self.batch_size)


loader = PracticeDataLoader("ai_practice.csv", batch_size=2)

for batch in loader:
    print(batch)
```

Reflection and exit ticket

1. Why must __iter__() create a fresh pipeline?
__iter__() needs to create a fresh pipeline because generators can only be used once. After the data has been read, the iterator is finished. Creating a new pipeline allows the data loader to be reused for multiple epochs.

2. What remains in memory while one batch is processed?
Only the current batch and a few temporary values from the pipeline are kept in memory. The whole dataset does not have to be loaded, which makes the process more efficient.

3. Why does list(generator) remove the main memory advantage?
Calling list(generator) forces the generator to produce every item immediately and store them all in a list. This means the entire dataset is loaded into memory, so the benefit of lazy loading is lost.

4. What cleanup does the file context manager guarantee?
A file context manager automatically closes the file after it is finished being used. This happens even if an error occurs, so it helps prevent files from being left open.

5. How could buffer-based approximate shuffling be added?
A small buffer could be used to hold a limited number of records at a time. The program could randomly choose items from the buffer while adding new records as it goes. This would create a shuffled order without needing to store the entire dataset.
